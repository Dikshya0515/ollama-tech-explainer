```python
# Imports
import ollama
```

```python
# Constants
MODEL_LLAMA = 'llama3.2'
```

```python
# Here is the question; type over this to ask something new
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""

print("Question:")
print(question)
```

```python
# Get Llama 3.2 to answer with streaming
print("\n" + "="*70)
print("ü§ñ Llama 3.2 Response:")
print("="*70 + "\n")

try:
    response = ollama.chat(
        model=MODEL_LLAMA,
        messages=[
            {
                'role': 'system',
                'content': 'You are a helpful coding instructor who explains code clearly and concisely.'
            },
            {
                'role': 'user',
                'content': question
            }
        ],
        stream=True  # Enable streaming for real-time response
    )
    
    # Print the response as it streams in
    for chunk in response:
        print(chunk['message']['content'], end='', flush=True)
    
    print("\n\n" + "="*70)
    
except Exception as e:
    print(f"‚ùå Error: {e}")
    print("\nMake sure Ollama is running and you've pulled the llama3.2 model!")
    print("Run in terminal: ollama pull llama3.2")
```
